```python
import pandas as pd

filename ="./adult.data"
census = pd.read_csv(filename)
census.head()
```

```python
column_name = ['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'martial-status', 'occupation', 'relationship',
              'race', 'sex','capital-gain', 'capital-loss', 'hours-per-week', 'native-country', 'Income']
```

```python
census = pd.DataFrame(census.values, columns=column_name)
census.head()
```

```python
print("Rows : " , census.shape[0])
print("Columns : " , census.shape[1])
print("Features : " , census.columns.tolist())
print("Missing values : " , census.isnull().sum().values.sum())
print("unique : " , census.nunique())
```

```python
census.info()
```

```python
census = census.infer_objects()
census.head()
```

```python
census.info()
```

```python
census['Income'] = census['Income'].str.strip()
census.tail()
```

```python
n_records = census.shape[0]
n_great_50k = census.loc[census['Income']  == '>50K'].shape[0]
n_at_most_50k = census[census['Income'] == '<=50K'].shape[0]
great_percentage = (n_great_50k/ n_records) * 100

print(n_records)
print(n_great_50k)
print(n_at_most_50k)
print(great_percentage)
```

```python
#'age', 'workclass', 'fnlwgt', 'education', 'education-num', 'martial-status', 'occupation', 'relationship',
            #  'race', 'sex','capital-gain', 'capital-loss', 'hours-per-week', 'native-country', 'Income'
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
numerical = ['age', 'education-num','capital-gain', 'capital-loss', 'hours-per-week']
cenus_minmax_transform = pd.DataFrame(data=census)
cenus_minmax_transform[numerical] = scaler.fit_transform(cenus_minmax_transform[numerical])
cenus_minmax_transform.head()
```

```python
income_raw = cenus_minmax_transform['Income']
cenus_minmax_transform = cenus_minmax_transform.drop('Income', axis=1)
```

```python
feature_final = pd.get_dummies(cenus_minmax_transform)

from sklearn.preprocessing import LabelEncoder
enconder = LabelEncoder()
income = income_raw.apply(lambda x:0 if x == "<=50K" else 1)
income = pd.Series(enconder.fit_transform(income_raw))
encoded = list(feature_final.columns)
print("{} total feature after one-hot enconding ".format(len(encoded)))
print(encoded)
```
