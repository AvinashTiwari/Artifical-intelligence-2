```python
#'age', 'workclass', 'fnlwgt', 'education', 'education-num', 'martial-status', 'occupation', 'relationship',
            #  'race', 'sex','capital-gain', 'capital-loss', 'hours-per-week', 'native-country', 'Income'
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
numerical = ['age', 'education-num','capital-gain', 'capital-loss', 'hours-per-week']
cenus_minmax_transform = pd.DataFrame(data=census)
cenus_minmax_transform[numerical] = scaler.fit_transform(cenus_minmax_transform[numerical])
cenus_minmax_transform.head()
```

```python
income_raw = cenus_minmax_transform['Income']
cenus_minmax_transform = cenus_minmax_transform.drop('Income', axis=1)
```

```python
feature_final = pd.get_dummies(cenus_minmax_transform)

from sklearn.preprocessing import LabelEncoder
enconder = LabelEncoder()
income = income_raw.apply(lambda x:0 if x == "<=50K" else 1)
income = pd.Series(enconder.fit_transform(income_raw))
encoded = list(feature_final.columns)
print("{} total feature after one-hot enconding ".format(len(encoded)))
print(encoded)
```

```python
cenus_minmax_transform.nunique()
```

```python
from sklearn.model_selection import train_test_split
import time

X_train,X_test,y_train,y_test = train_test_split(feature_final, income, test_size=0.2,random_state=0)
print("Training set has {} samples".format(X_train.shape[0]))
print("Test set has {} samples".format(X_test.shape[0]))
```

```python
accuracy = n_great_50k / n_records
precision = n_great_50k / (n_great_50k  + n_at_most_50k)
recall = n_great_50k / (n_great_50k + 0)
fscore = (1 + (0.5 *0.5)) * (precision * recall / ((0.5*0.5 * (precision)) + recall))
print(accuracy)
print(fscore)
```

```python
from sklearn.metrics import fbeta_score, accuracy_score
from time import time
def train_predict(learner, sample_size, X_train, y_train, X_test, y_test):
    results = {}
    start = time()
    learner = learner.fit(X_train[:sample_size], y_train[:sample_size])
    end = time()
    
    results['train_time'] = end - start
    
    start = time()
    predictions_test = learner.predict(X_test)
    predictions_train = learner.predict(X_train[:300])
    end = time()
    results['pred_time'] = end - start
    
    
    results['acc_train'] = accuracy_score(y_train[:300], predictions_train)
    results['acc_test'] = accuracy_score(y_test, predictions_test)
    results['f_train'] = fbeta_score(y_train[:300], predictions_train,0.5)
    results['f_test'] = fbeta_score(y_test, predictions_test,0.5)
    print("{} trained on {} sample ".format(learner.__class__.__name__, sample_size))
    return results
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import AdaBoostClassifier

clf_A = DecisionTreeClassifier(random_state=101)
clf_B = SVC(random_state=101)
clf_C = AdaBoostClassifier(random_state=101)

sample_1 = int(round(len(X_train)/100))
sample_10 = int(round(len(X_train)/100))
sample_100 = len(X_train)

results = {}

for clf in [clf_A, clf_B, clf_C]:
    clf_name = clf.__class__.__name__
    results[clf_name] = {}
    for i , samples in enumerate([sample_1,sample_10,sample_100]):
        results[clf_name][i] = \
        train_predict(clf, samples, X_train, y_train, X_test, y_test)
        
```

```python
for i in results.items():
    print(i[0])
    display(pd.DataFrame(i[1]).rename(columns={0:'1%', 1:"10%", 2:"100%"}))
```

```python
from sklearn.metrics import confusion_matrix
plt.figure(figsize=(30,12))
for i, model in enumerate([clf_A, clf_B,clf_C]):
    cm = confusion_matrix(y_test, model.predict(X_test))
    cm = cm.astype('float') / cm.sum(axis=1)[:,np.newaxis]
    
    plt.figure(i)
    sns.heatmap(cm, annot=True, annot_kws={"size" :10} , cmap='Blues',square=True, fmt='.3f')
    plt.ylabel('True label')
    plt.xlabel('Predict label')
    plt.title('Confusion matrix for :\n {}'.format(model.__class__.__name__))
```
