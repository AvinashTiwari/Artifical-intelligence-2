```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('./WA_Fn-UseC_-Marketing-Campaign-Eff-UseC_-FastF.csv')
df.head()
```

```python
print("Rows : ", df.shape[0])
print("Columns : ", df.shape[1])
print("Feature : ", df.columns.tolist())
print("Missing values : ", df.isnull().sum().values.sum())
print("Unique values : ", df.nunique())
```

```python
df.describe()
```

```python
plt.style.use('ggplot')
ax = df.groupby('Promotion').sum()['SalesInThousands'].plot.pie(figsize=(8,8),
                                                      autopct='%1.0f%%',
                                                      shadow=True,
                                                      explode = (0,0.1,0))
ax.set_ylabel(' ')
ax.set_title('Sales Distribution Across the 3 different Promotion')
plt.show()
```

```python
df.groupby(['Promotion', 'MarketSize']).count()['MarketID']
```

```python
ax = df.groupby(['Promotion', 'MarketSize']).count()['MarketID'].unstack('MarketSize').plot(
            kind='bar',
            figsize=(12,10),
            grid=True)
ax.set_ylabel('count')
ax.set_title('breakdown of market sizes across different promotion')
plt.show()
```

```python
ax = df.groupby(['Promotion', 'MarketSize']).count()['MarketID'].unstack('MarketSize').plot(
            kind='bar',
            figsize=(12,10),
            grid=True,
            stacked=True)
ax.set_ylabel('count')
ax.set_title('breakdown of market sizes across different promotion')
plt.show()
```

```python
ax = df.groupby('AgeOfStore').count()['MarketID'].plot(
            kind='bar',
            figsize=(12,7),
            grid=True)
ax.set_ylabel('age')
ax.set_xlabel('count')

ax.set_title('Overall Dist Store Age')
plt.show()
```

```python
df.groupby(['AgeOfStore', 'Promotion']).count()['MarketID']
```

```python
df.groupby(['AgeOfStore', 'Promotion']).count()['MarketID'].unstack('Promotion').iloc[::-1].plot(
            kind='barh',
            figsize=(14,18),
            grid=True)
ax.set_ylabel('age')
ax.set_ylabel('count')

ax.set_title('Overall dist of age store')
plt.show()
```
