```python
import matplotlib.pyplot as plt
import pandas as pd
file_name = "./bank-full.csv"
bank = pd.read_csv(file_name, sep=";")
bank.head()
```

```python
print("Rows ", bank.shape[0])
print("Columns ", bank.shape[1])
print("Feature ", bank.columns.tolist())
print("Missing Value ", bank.isnull().sum().values.sum())
print("Unique value  ", bank.nunique())


```

```python
bank.describe()
```

```python
bank.info()
```

```python
bank['converted'] = bank['y'].apply(lambda x: 0 if x == 'no' else 1)
del bank['y']
bank.head()
```

```python
ax = bank[['converted', 'balance']].boxplot(by='converted', showfliers=True, figsize=(10,7))

ax.set_xlabel('Conversion')
ax.set_ylabel('Average Bank Balance')
ax.set_title('Average Bank balance Dist Conversion')

plt.suptitle("")
plt.show()
```

```python
ax = bank[['converted', 'balance']].boxplot(by='converted', showfliers=False, figsize=(10,7))

ax.set_xlabel('Conversion')
ax.set_ylabel('Average Bank Balance')
ax.set_title('Average Bank balance Dist Conversion')

plt.suptitle("")
plt.show()
```

```python
import seaborn as sns
fontsize = 10
fig, axes = plt.subplots()
fig.set_size_inches(10,8)
sns.violinplot('converted', 'balance', data=bank , ax = axes)
axes.set_title('Average bank balace dist by conversion')
axes.yaxis.grid(True)
axes.set_xlabel('Conversion')
axes.set_ylabel('Average Bank Balance')
plt.show()
```

```python
bank['campaign'].unique()
```

```python
conversion_by_contact = bank.groupby('campaign')['converted'].sum() / bank.groupby('campaign')['converted'].count() *100.0
conversion_by_contact.head(10)
```

```python
ax =  conversion_by_contact.plot(kind='bar',
                                figsize=(10,8),
                                title="Conversion Rates by Number of Contract",
                                grid=True,
                                color='coral')
ax.set_xlabel('Number of Contact')
ax.set_ylabel('Conversion Rate(%)')
plt.show()
```

```python
conversion_by_job = bank.groupby('job')['converted'].sum() / bank.groupby('job')['converted'].count() *100.0
conversion_by_job.head(10)
```

```python
ax =  conversion_by_job.plot(kind='barh',
                                figsize=(10,8),
                                title="Conversion Rates by Number of Job",
                                grid=True,
                                color='coral')
ax.set_xlabel('Number of Job')
ax.set_ylabel('Conversion Rate(%)')
plt.show()
```

```python
Y_train = bank['converted']
X_train = bank.drop(labels = ['converted'], axis=1)
X_train
```

```python
from sklearn import tree
dec_tree_model = tree.DecisionTreeClassifier(max_depth=3)
```

```python
dec_tree_model.fit(X_train, Y_train)
```
