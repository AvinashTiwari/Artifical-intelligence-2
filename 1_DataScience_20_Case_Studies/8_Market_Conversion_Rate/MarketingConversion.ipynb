```python
import pandas as pd
file_name = "./bank-additional-full.csv"
bank = pd.read_csv(file_name, sep=";")
```

```python
bank.head()
```

```python
print("Rows ", bank.shape[0])
print("Columns ", bank.shape[1])
print("Feature ", bank.columns.tolist())
print("Missing value ", bank.isnull().sum().values.sum())
print("Unique value ", bank.nunique())
```

```python
bank['converted'] = bank['y'].apply(lambda c:1 if c =='yes' else 0)
bank =  bank.drop(['y'], axis=1)
```

```python
import matplotlib.pyplot as plt

def plot_corr(df, size=10):
    corr = df.corr()
    fig, ax = plt.subplots(figsize=(size,size))
    ax.legend()
    cax = ax.matshow(corr)
    fig.colorbar(cax)
    plt.xticks(range(len(corr.columns)),corr.columns, rotation='vertical' )
    plt.yticks(range(len(corr.columns)),corr.columns )
    plt.show()
    
    
plot_corr(bank)
    
```

```python
print('Total converted : %i out of %i' %(bank.converted.sum(), len(bank)))
```

```python
print('Our Conversion Rate  %0.2f%%' %(bank.converted.sum()/ len(bank) * 100.0))
```

```python
pd.DataFrame(bank.groupby(by="campaign")['converted'].sum())
```

```python
pd.DataFrame(bank.groupby(by="campaign")['converted'].count())
```

```python
conv_by_camp = bank.groupby('campaign')['converted'].sum()/bank.groupby('campaign')['converted'].count() * 100.0
conv_by_camp.head()
```

```python
converted_sums_camp = bank.groupby(['campaign', 'converted']).agg({'converted' : 'count'})
converted_sums_camp2 = converted_sums_camp.groupby(level=0).apply(lambda x: 100 *x / float(x.sum()))
converted_sums_camp2
```

```python
converted_camp_age = bank.groupby(['age', 'converted']).agg({'converted' : 'count'})
conv_by_age = converted_camp_age.groupby(level=0).apply(lambda x: 100 *x / float(x.sum()))
conv_by_age
```
